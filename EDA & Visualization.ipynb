```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Processed')
```

```python
df
```

```python
sns.countplot(x='addicted',data=df,palette=['#4C72B0', '#DD8452'])
plt.title("Traget Variable Distribution")
plt.xlabel(' Not-addicted (0) and Addicted (1) ')
plt.legend(['No','Yes'])
plt.show()
```

```python
sns.countplot(hue='addicted', x='Gender?', data=df,palette=['#4C72B0', '#DD8452'])
plt.title("Gender-wise Addiction")
plt.xlabel("Addicted")
plt.ylabel("Average Gaming Hours")

plt.show()
```

```python
sns.barplot(x='addicted', y='The average time you spend playing games?', data=df,palette=['#4C72B0', '#DD8452'])
plt.title("Average Gaming Hours by Addiction")
plt.xlabel("Addicted")
plt.ylabel("Average Gaming Hours")

plt.show()
```

```python

sns.countplot(x='When you go to sleep?',hue='addicted', data=df,palette=['#4C72B0', '#DD8452'])
plt.title("Average Gaming Hours by Addiction")
plt.xlabel("Addicted (0 = No, 1 = Yes)")
plt.ylabel("Average Gaming Hours")
plt.show()
```

```python
sns.countplot(x='Which type of game you addict more?',hue='addicted',data=df)
plt.title("Game Type and Addiction")
plt.xlabel("Type of Game")
plt.legend(['No','Yes'])
plt.show()
```
